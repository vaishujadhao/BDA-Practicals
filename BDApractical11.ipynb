```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
print(x)
```

```python
y=x*2
print(y)
```

```python
plt.plot(x,y)
plt.title("Line Chart")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```

```python
plt.bar(x,y)
plt.title("Line Chart")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```

```python
plt.scatter(x,y)
plt.title("Line Chart")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```
